```python
import cv2
import os
from dotenv import load_dotenv

load_dotenv()

ML_Summer_School_ID = os.getenv('ML_Summer_School_ID')
print("Your Sudent ID is: " + ML_Summer_School_ID)
```

```python
import random
import matplotlib.pyplot as plt

test_img_path = "./test_dataset/sample_data"
os.listdir(test_img_path)
```

```python
selected_img = random.choice(os.listdir(test_img_path))
print("selected_img : " + selected_img)



selected_img_full_path = test_img_path+"/"+selected_img

print("selected_img_full_path : ",selected_img_full_path)

read_img = cv2.imread(selected_img_full_path)
if read_img is None:
    print("Image not found")


resize_img = cv2.resize(read_img,(300,300))
plt.imshow(resize_img)
```

```python
# Write your image into data_annotated folder
write_name = f"./data_annotated/{ML_Summer_School_ID}.png"
cv2.imwrite(write_name, resize_img)
```
